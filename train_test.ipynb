```python
import pandas as pd

# Load training datasets
emotion_df = pd.read_csv(r'C:\Users\ajayn\Desktop\lg\twitter_analytics_task\data\emotions_train.csv')
sentiment_df = pd.read_csv(r'C:\Users\ajayn\Desktop\lg\twitter_analytics_task\data\training_senti.csv')

# Load final dataset
tweet_df = pd.read_pickle(r'C:\Users\ajayn\Desktop\lg\twitter_analytics_task\data\twitter_final_extract_cadmv.p')

print("Emotion data:", emotion_df.shape)
print("Sentiment data:", sentiment_df.shape)
print("Target data:", tweet_df.shape)
```

```python
import re
import string

def clean_tweet(text):
    text = re.sub(r'http\S+', '', text)  # remove URLs
    text = re.sub(r'@\w+', '', text)     # remove mentions
    text = re.sub(r'#\w+', '', text)     # remove hashtags
    text = re.sub(r'\d+', '', text)      # remove numbers
    text = text.lower()
    text = re.sub(f"[{re.escape(string.punctuation)}]", '', text)
    return text.strip()
```

```python
# Clean emotion and sentiment training data
emotion_df['clean_text'] = emotion_df['content'].apply(clean_tweet)
sentiment_df['clean_text'] = sentiment_df['tweet'].apply(clean_tweet)

# Clean target data
tweet_df['clean_text'] = tweet_df['text'].apply(clean_tweet)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Features and labels
X_emotion = emotion_df['clean_text']
y_emotion = emotion_df['sentiment']

# Vectorize
tfidf = TfidfVectorizer(max_features=5000)
X_emotion_vec = tfidf.fit_transform(X_emotion)

# Train-test split
X_train_e, X_test_e, y_train_e, y_test_e = train_test_split(X_emotion_vec, y_emotion, test_size=0.2, random_state=42)

# Train model
emotion_model = LogisticRegression(max_iter=200)
emotion_model.fit(X_train_e, y_train_e)

# Evaluate
y_pred_e = emotion_model.predict(X_test_e)
print(classification_report(y_test_e, y_pred_e))
```

```python
# Features and labels
X_senti = sentiment_df['clean_text']
y_senti = sentiment_df['polarity'].replace(4, 1)  # 0 = negative, 1 = positive

# Reuse same TF-IDF vectorizer for consistency
X_senti_vec = tfidf.transform(X_senti)

# Train-test split
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_senti_vec, y_senti, test_size=0.2, random_state=42)

# Train model
senti_model = LogisticRegression(max_iter=200)
senti_model.fit(X_train_s, y_train_s)

# Evaluate
y_pred_s = senti_model.predict(X_test_s)
print(classification_report(y_test_s, y_pred_s))
```

```python
# Vectorize final tweets
tweet_vec = tfidf.transform(tweet_df['clean_text'])

# Predict emotion & sentiment
tweet_df['predicted_emotion'] = emotion_model.predict(tweet_vec)
tweet_df['predicted_sentiment'] = senti_model.predict(tweet_vec)
```

```python
import yake

kw_extractor = yake.KeywordExtractor(top=1)

def extract_topic(text):
    try:
        keywords = kw_extractor.extract_keywords(text)
        return keywords[0][0] if keywords else ''
    except:
        return ''

tweet_df['topic'] = tweet_df['clean_text'].apply(extract_topic)

```

```python
avg_polarity_per_topic = tweet_df.groupby('topic')['predicted_sentiment'].mean().sort_values(ascending=False)

print("🔹 Average polarity per topic:\n")
print(avg_polarity_per_topic.head(10))
```

```python
top_emotions_per_topic = (
    tweet_df.groupby('topic')['predicted_emotion']
    .value_counts()
    .groupby(level=0)
    .nlargest(3)
    .reset_index(level=0)
)

```

```python
print("🔹 Top 3 emotions per topic:\n")
top_emotions_per_topic.head(10)
```

Overall Average Sentiment of Most Popular Tweets

```python
tweet_df['popularity'] = tweet_df['favorite_count'] + tweet_df['retweets_count']

# Sort by popularity and take top 50
popular_tweets = tweet_df.sort_values(by='popularity', ascending=False).head(50)

# Get average sentiment (0 = negative, 1 = positive)
popular_avg_sentiment = popular_tweets['predicted_sentiment'].mean()

print(f"🔹 Overall average sentiment of most popular tweets: {popular_avg_sentiment:.2f}")
```
